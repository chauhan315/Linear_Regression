```python
data.Explicit = data.Explicit.map(lambda x: 1 if x == True else 0)
```

```python
ef_var = ['Duration (ms)', 'Explicit', 'Danceability', 'Energy', 'Loudness', 'Acousticness', 'Popularity']
```

```python
data_new = data[ef_var]
data_new
```

# Model Building

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV, KFold

X = data_new.drop('Popularity', axis=1, inplace=False)
y = data.Popularity

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=47)

model = Pipeline([
    ("scaler", StandardScaler()),
    ("rf_rg", RandomForestRegressor())
])
```

```python
model.fit(X_train, y_train)
```

#### hyperparameter tuning

```python
param_grid = {
    'rf_rg__n_estimators': [50, 100, 200],
    'rf_rg__max_features': ['auto', 'sqrt', 'log2'],
    'rf_rg__max_depth': [10, 20, 30, None],
    'rf_rg__min_samples_split': [2, 5, 10],
    'rf_rg__min_samples_leaf': [1, 2, 4]
}
kfolds = 5

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=kfolds, n_jobs=-1, 
                    return_train_score=True, scoring='r2', verbose=True)
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
model_best = Pipeline([
    ("scaler", StandardScaler()),
    ("rf_rg", RandomForestRegressor(max_depth=10, max_features='log2', min_samples_leaf=2,
                      min_samples_split=10))
])
model_best.fit(X_train, y_train)
```

```python
model_lst = [model, model_best]
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
for i, m in enumerate(model_lst):
    y_pred = m.predict(X_test)
    print('Model', i)
    print(r2_score(y_test, y_pred))
    print(mean_absolute_error(y_test, y_pred))
    print(mean_squared_error(y_test, y_pred))
    print("-"*50)
```
